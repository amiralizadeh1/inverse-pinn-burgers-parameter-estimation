```python

def calculate_variance(numbers):
    # Check if list is empty
    if not numbers:
        return 0
    
    # Calculate mean
    mean = sum(numbers) / len(numbers)
    
    # Calculate sum of squared differences from mean
    squared_diff_sum = sum((x - mean) ** 2 for x in numbers)
    
    # Calculate variance by dividing by n
    variance = squared_diff_sum / len(numbers)
    
    return variance

# Example usage:
P1_adam = [0.743192, 0.927746, 0.842724]
P2_adam = [0.867293, 0.870294, 0.851214]
P3_adam = [1.090215, 1.235744, 1.02028]
P4_adam = [1.313944, 0.948416, 1.255720]
P5_adam = [0.903507, 0.814402, 0.923963]
P1_powell = [2.987571, 4.835951, 1.955139]
P2_powell = [0.946170, 0.941387, 0.967138]
P3_powell = [2.012354, 1.124724, 4.762250]
P4_powell = [1.007508, 0.998632, 1.025746]
P5_powell = [0.000166, 0.000232, 0.001369]
# Calculate and print variance for each list
var_P1_adam = calculate_variance(P1_adam)
var_P2_adam = calculate_variance(P2_adam)
var_P3_adam = calculate_variance(P3_adam)
var_P4_adam = calculate_variance(P4_adam)
var_P5_adam = calculate_variance(P5_adam)

var_P1_powell = calculate_variance(P1_powell)
var_P2_powell = calculate_variance(P2_powell)
var_P3_powell = calculate_variance(P3_powell)
var_P4_powell = calculate_variance(P4_powell)
var_P5_powell = calculate_variance(P5_powell)

print(f"Variance P1 Adam: {var_P1_adam}")
print(f"Variance P2 Adam: {var_P2_adam}")
print(f"Variance P3 Adam: {var_P3_adam}")
print(f"Variance P4 Adam: {var_P4_adam}")
print(f"Variance P5 Adam: {var_P5_adam}")

print(f"Variance P1 Powell: {var_P1_powell}")
print(f"Variance P2 Powell: {var_P2_powell}")
print(f"Variance P3 Powell: {var_P3_powell}")
print(f"Variance P4 Powell: {var_P4_powell}")
print(f"Variance P5 Powell: {var_P5_powell}")
```
